# SORTING

Sorting is the process of arranging data in a specific order, such as ascending (small → big) or descending (big → small).

Example:

```python
# Ascending order
numbers = [5, 2, 9, 1, 7]

numbers.sort()
print(numbers)
```

```python
# Descending order
numbers.sort(reverse=True)
print(numbers)
```

### argsort()

argsort() returns the indices that would sort an array.

Example:

```python
a = np.array([40, 10, 30, 20])

result = np.argsort(a)
print(result)
```

### where()

where() is used to find indices where a condition is true, or to choose values based on a condition

Example:

```python
a = np.array([2, 6, 4, 9, 1])

result = np.where(a > 5)
print(result)
```

```python
# Replace values using condition
a = np.array([2, 6, 4, 9, 1])

result = np.where(a > 5, 1, 0)
print(result)
```

# SEARCHING

Searching is the process of finding a specific element in a collection of data.

Example:

```python
numbers = [10, 20, 30, 40, 50]

if 30 in numbers:
    print("Element found")
else:
    print("Element not found")
```

```python
# Searching using loop
key = 40
found = False

for num in numbers:
    if num == key:
        found = True
        break

print(found)
```
