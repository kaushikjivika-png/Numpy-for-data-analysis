### Introduction of NumPy


- Numpy is a Python Library used to perform Mathematical operations.
- It is used to work with arrays.
- It also has functions for working in domain of Linear Algebra and Matrices.

### Why NumPy is used?

- In Python we have List that serves the purpose of NumPy, but they are slow to process.
- NumPy aims to provide an array object that is upto 50x faster than the Python List.
- Arrays are frequently used in Data Science and Data Analysis, where speed and resource are very important.
