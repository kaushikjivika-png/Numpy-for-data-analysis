Question 34: Sort an array.

Solution:

```python
import numpy as np

arr_1 = np.array([12,3,5,43,67])
sort = np.sort(arr_1)
print(sort)
```

Question 35: Sort a 2D array row-wise.

Solution:

```python
arr_2D = np.array([[12,3,45,3,4],
                   [12,34,6,21,6]])
sort_2D = np.sort(arr_2D)
print(sort_2D)
```

Question 36: Find index of max value.

Solution:

```python
arr_max = np.array([12,3,4,21,4,5])
print(np.argmax(arr_max))
```

Question 37: Use where() condition.

Solution:

```python
array = np.array([12,3,45,32,6])
print(np.where(array == 45))
```
