# Statistical Operations

- MAX -> To find maximum value.                       -> np.max() 
- MIN -> To find minimum value.                       -> np.min() 
- SQRT -> To find square root.                        -> np.sqrt()
- ARGMIN -> To find min value with index no.          -> np.argmin()
- ARGMAX -> To find max value with index no.          -> np.argmax
- CUMSUM -> To find cumsum of array.                  -> np.cumsum()
- SIN -> To find sin value of array.                  -> np.sin()
- COS -> To find cos value of array.                  -> np.cos()

Example:

```python
# MAXIMUM VALUE / ARGMAX:

arr = np.array([1,23,45,65,34])
print("MAXIMUM VALUE : ",np.max(arr))
print("MAXIMUM VALUE AT INDEX : ",np.argmax(arr))


# MINIMUM VALUE / ARGMIN:

arr1 = np.array([1,23,45,65,34])
print("MINIMUM VALUE : ",np.min(arr1))
print("MINIMUM VALUE AT INDEX : ",np.argmin(arr1))


# SQUARE ROOT:

arr2 = np.array([1,4,16,25,36])
print("SQUARE ROOT : ",np.sqrt(arr2))

# CUMSUM:

arr3 = np.array([1,23,45,65,34])
print("CUMSUM VALUE : ",np.cumsum(arr3))

# SIN VALUE:

arr4 = np.array([1,23,45,65,34])
print("SIN VALUE : ",np.cumsum(arr4))




```
