# Aggregation Functions

This notebook covers:

   - Row-wise operations
   - Column-wise operations
   - Axis concept (axis=0, axis=1)

ðŸ“Œ What are Aggregation Functions?

Aggregation functions are used to summarize or combine multiple values into a single result.



### Row-wise operation

Row-wise operations perform calculations across rows.

Each row is treated independently.

Example:

```python
import numpy as np

arr = np.array([
    [10, 20, 30],
    [40, 50, 60]
])
np.sum(arr, axis=1)
np.mean(arr, axis=1)

```

### Column-wise operation 

Column-wise operations perform calculations down the columns.

Each column is treated independently.

Example:

```python
arr = np.array([
    [10, 20, 30],
    [40, 50, 60]
])
np.sum(arr, axis=0)
np.mean(arr, axis=0)
```

### Axis-concept

ðŸ”¹ axis = 0 (Default)

-  Operation happens column-wise

-  Moves down the rows

ðŸ”¹ axis = 1

-  Operation happens row-wise

-  Moves across columns
