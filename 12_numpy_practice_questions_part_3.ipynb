Question 30: Convert 1D array to 2D.

Solution:

```python
import numpy as np

arr = np.array([12,34,5,21,3,5])
arr_2d = arr.reshape(2,3)
print(arr_2d)
```

Question 31: Convert 2D array into 1D.

Solution:

```python
arr_2d = np.array([[12,3,4,5]])
print(arr_2d.flatten())
```

Question 32: Convert matrix into column vector.

Solution:

```python
matrix = np.array([[12,34,2,5,6],
                   [1,7,5,4,2]])
new_matrix = matrix.reshape(-1,1)
print(new_matrix)
```

Question 33: Convert matrix into row vector.

Solution:

```python
matrix_2 = np.array([[12,3,2,4],
                     [3,4,5,6]])
new_matrix_2 = matrix_2.reshape(1,-1)
print(new_matrix_2)
```
